Importing the proper libraries.


```python
import pandas as pd
import numpy as np
```

Turing data into df.

```python
df = pd.read_csv('/content/train.csv')
df.info()
```

**Question 1** We first drop the null rows for Milage, Engine, Power, and Seats simply becuase, if we do so, we will only lose 40 out of nearly 6000 rows. We then take the units out of the new price, so we can convert to a float. We then replace that float with the median since, assuming our cars should be representative of the car population, the most common price seems like a good estimation.

```python
df.dropna(subset=['Mileage','Engine','Power','Seats'], inplace=True)
df['New_Price'] = df['New_Price'].astype(str).str.replace(' Lakh','').str.replace(' Cr','').astype(float)
df['New_Price'] = df['New_Price'].fillna(df['New_Price'].median())
df.info()
```

**Question 2** We already removed the text from New Price in question 1. The rest are removed below.

```python
df['Mileage'] = df['Mileage'].astype(str).str.replace(' kmpl','').str.replace(' km/kg','').astype(float)
df['Engine'] = df['Engine'].astype(str).str.replace(' CC','').astype(float)
df['Power'] = df['Power'].astype(str).str.replace(' bhp','').astype(float)
df
```

**Question 3** Here we convert Fuel_Type and Transmission to numerical values as the question asks.

```python
df['Fuel_Type'] = df['Fuel_Type'].astype('category').cat.codes
df['Transmission'] = df['Transmission'].astype('category').cat.codes
df
```

**Question 4** Here we define a new column that finds the difference between the cars the price and current price to see how much cheaper/expensive a car has become.

```python
df.assign(Change_In_Price = df['New_Price'] - df['Price'])
```

**Question 5** Below I perform the python equivilent of select, filter, rename, mutate, arrange, and summerize.

```python
new_df = df.loc[df['New_Price']== df['New_Price'].median()].rename(columns = {'Name' : "Make/Model"})[['Make/Model','Mileage']].sort_values(by='Mileage')
new_df.groupby('Make/Model').agg('mean')
```
